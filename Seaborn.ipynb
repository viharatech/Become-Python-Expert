```python
df.corr()
```

```python
sns.heatmap(df.corr(),vmin=0,vmax=1,center = 1 , annot=True)
```

### joint plot

joint plot comes under bivarite annalysis because  it has 2 aruguments X and y point

```python
df.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data = df,kind = 'hex',color = 'g')
```

### dist_plot

dist_plot is used to check the distribution od column features it is an univariate analysis

```python
sns.distplot(df['tip'],color='blue') # hist  ,  kde
```

```python
sns.distplot(df['total_bill'],hist=True,color='green',kde=True,bins=50)
```

```python
sns.distplot(df['size'],hist=True,color='green')
```
